## Import libraries

```
import numpy as np
import pandas as pd
#import vaex
```

## Create Dummy Data

```
a = np.random.rand(74, 3000)
a.shape
```

```
index = [f'R{i}' for i in range(a.shape[0])]
columns = [f'C{i}' for i in range(a.shape[1])]
dfp = pd.DataFrame(a, columns=columns, index=index)
dfp.head()
```

```
%%time
dfp.T.corr()
```

```
%%time
dfp_corr = dfp.corr()
```

```
target_cols = ['C0', 'C12', 'C100']
corr_result = dfp_corr[target_cols]
print(corr_result)
```

```
print(dfp_corr)
```
