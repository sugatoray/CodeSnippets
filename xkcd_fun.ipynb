````
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg' # 'svg', 'retina'
plt.style.use('seaborn-white')
````

````
# Make data
x = np.arange(-0.7, 7, 0.1)
y = 0.6*(x**2 - 8*x + 10)
````

## **Show Figure in _`xkcd`_-Style**

You can make a regular plot with matplotlib. But then to add some `xkcd`-type features to it, all you need is the `plt.xkcd()` context manager. Anything that goes inside the context manager will have the `xkcd`-formatted-layout.  

```python
with plt.xkcd():
    plt.plot(x,y)
    #... your codes goes here
```

````
with plt.xkcd():
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Demo: xkcd-style plot')
    #plt.legend()
    plt.show()
````

# **Show Image**

We will pull image data from an _image-url_, read it with `skimage.io.imread(url)` and then show it with `plt.imshow(image)` as follows.

````
image_url = r"https://s.ftcdn.net/v2013/pics/all/curated/RKyaEDwp8J7JKeZWQPuOVWvkUjGQfpCx_cover_580.jpg?r=1a0fc22192d0c808b8bb2b9bcfbf4a45b1793687"
````

````
#!pip install scikit-image
from skimage import io
#import matplotlib.pyplot as plt

# create a file-like object from the url
url = image_url

# read the image file in a numpy array
image = io.imread(url)
plt.imshow(image)
plt.show()
````

## **Show Image in _`xkcd`_-Style**

````
with plt.xkcd():
    plt.imshow(image)
    #plt.gca().set_axis_off()
    plt.yticks(ticks=[]) # set yticks off
    plt.xticks(ticks=[]) # set xticks off
    plt.title('xkcd-Styled Image Display')
    plt.show()
````
